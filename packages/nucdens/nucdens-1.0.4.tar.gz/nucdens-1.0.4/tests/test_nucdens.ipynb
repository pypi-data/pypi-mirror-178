```python
import os
import sys
import shutil
import pandas as pd 
# to see all column of pandas table 
pd.set_option('display.max_columns', None)
```

```python
from nucdens import access
# create a test library locally using the standard web location for density files
testdf=access.database(workdir=os.environ["HOME"]+"/work/denstest")
```

```python
# print current table and the file name for the HDF file of the table 
print("filename of table: ",testdf.pddf_hdf)
testdf.pddf
```

```python
# get file of line 36
row=testdf.pddf.loc[[36]]
densid=row.to_dict('records')[0]
hashname,uniquename=testdf.get_file(**densid)
print("hashname:    ",hashname)
print("filesize:    ",os.path.getsize(hashname))
print("uniquename:  ",uniquename)

# read in for testing content and compress 
density2b=access.densfile2b(hashname,printlevel=2)
density2b.compress(hashname,1e-6)

# copy to new location using the unique file name 
shutil.copy(hashname,os.environ["HOME"]+"/work/denstest/"+uniquename) 

# get file of line 4
row=testdf.pddf.loc[[4]]
densid=row.to_dict('records')[0]
hashname,uniquename=testdf.get_file(**densid)
print("hashname:    ",hashname)
print("filesize:    ",os.path.getsize(hashname))
print("uniquename:  ",uniquename)

density1b=access.densfile1b(hashname,printlevel=2)

# copy to new location using the unique file name  
shutil.copy(hashname,os.environ["HOME"]+"/work/denstest/"+uniquename) 

print(os.listdir(os.environ["HOME"]+"/work/denstest"))
```

```python
# also cleanup new directory 
testdf.cleanup()
print(os.listdir(os.environ["HOME"]+"/work/denstest"))
```
